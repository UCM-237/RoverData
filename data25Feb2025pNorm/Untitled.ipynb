```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(x):
    return (4.0)*x*(1-x)

Niter = 100
x0 = 0.1
x02 = 0.1000001
xx1 = np.zeros(Niter)
xx2 = np.zeros(Niter)
xx1[0] = x0
xx2[0] = x02
for k in range(Niter-1):
    xx1[k+1] = logistic(xx1[k])
    xx2[k+1] = logistic(xx2[k])
    #xx2[k+1] = logistic(logistic(xx2[k]))

plt.figure(figsize=(18,4))
#plt.stairs(xx1,color='b')
plt.plot(xx1,'b')
plt.plot(xx2, 'r')
plt.plot(np.abs(xx2-xx1))
#plt.stairs(xx2,color='r')
#plt.plot(xx2,'r.')
```

```python
nInit = 1000
nIter = 1000

def funG(x):
    if(x <= 0.5):
        return 2 * x
    else:
        return 2 - 2 * x
        
xx = np.zeros(nIter)
xxF = np.zeros(nInit)
for k in range(nInit):
    xx[0] = np.random.rand()
    for j in range(nIter-1):
        xx[j+1] = funG(xx[j])
    if(xx[-1] >= 0 and xx[-1] <= 1):
        xxF[k] = xx[-1]

plt.plot(xxF, np.zeros(nInit), 'k.', markersize=1)
```
